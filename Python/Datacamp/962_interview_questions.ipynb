```python
list1 = ["a", '1', "b", "c", '3', '5']

def plusnum(list1, total=0):
    if len(list1) == 0:
        return total
    item = list1.pop(0)
    try:
        if type(int(item))==int:
            total += int(item)   
    except:
        pass
    
    return plusnum(list1, total) 

print(plusnum(list1))
```

```python
from collections import defaultdict

def group_by_owners(files):
    file = defaultdict(list)
    for key, value in files.items():
        file[value].append(key)
    return dict(file)

if __name__ == "__main__":    
    files = {
        'Input.txt': 'Randy',
        'Code.py': 'Stan',
        'Output.txt': 'Randy'
    }   
    print(group_by_owners(files))
```

```python
class IceCreamMachine:
    
    def __init__(self, ingredients, toppings):
        self.ingredients = ingredients
        self.toppings = toppings
        
    def scoops(self):
        liste = []
        for item in self.ingredients:
            for item2 in self.toppings:
                liste.append([item, item2])
        return liste
    
if __name__ == "__main__":
    machine = IceCreamMachine(["vanilla", "chocolate"], ["chocolate sauce"])
    print(machine.scoops())
```

## Prime Number Function

```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**(1/2))+1):
        if n%i==0:
            return False
    return True

print(is_prime(59))
```

## GCD  (Greatest Common Divisor) and Coprime number sequence 
For example, the numbers 7 and 9 are coprime because their GCD is 1.

```python
def gcd(a,b):
    if a<0 or b<0:
        print('You must enter a pozitive number')
    if a==0 or b==0:
        return 0
    if a==1 or b==1:
        return 1
    while b!=0:
        temp_a = a
        a=b
        b = temp_a%b
    return a
    
if gcd(7, 9)==1:
    print("They are coprime because GCD is 1.")  
```

## Throw a dice
Generator is a special iterable created by function having yield keyword in its body.
Generator is both iterable and iterator.

```python
from random import randint
def simulate_dice_throws():
    total, out = 0, dict([(i, [0, 0]) for i in range(1, 7)])
    while True:
        # Simulate a single toss to get a new number
        num = randint(1, 6)
        total += 1
        # Update the number and the ratio of realizations
        out[num][0] = out[num][0] + 1
        for j in range(1, 7):
        	out[j][1] = round(out[j][0]/total, 2)
        # Yield the updated dictionary
        yield out

# Create the generator and simulate 1000 tosses
dice_simulator = simulate_dice_throws()
for i in range(1, 1001):
    print(str(i) + ': ' + str(next(dice_simulator)))
```

 As you can see from the output, the more you throw a dice, the more the ratios become equal. These values can be seen as probabilities of a single realization.

## Recursion

#### factorial
n! = n * (n-1)!

```python
def fact(n):
    if n==1:
        return 1
    return n*fact(n-1)
fact(4)
```

#### Calculate an average value
avg = ( x(n-1) + (n-1) * avg(:-1) ) / n

```python
# Calculate an average value of the sequence of numbers
def average(nums):
    n = len(nums)
    # Base case
    if len(nums) == 1:  
        return nums[0]
    
    # Recursive call
    
    return (nums[n-1] + (n - 1) * average(nums[:-1])) / n  

# Testing the function
print(average([1, 2, 3, 4, 5]))
```
